# K Means

```python
import numpy as np
import matplotlib.pyplot as plot
import pandas as pd
from sklearn.cluster import KMeans
```

```python
g_datas = g_data  #.reshape(-1,1)
```

```python
g_datas
```

```python
#Dimensionality Reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(g_datas)
pca_2d = pca.transform(g_datas)

# finding optimal clusters
array= []
for i in range(1,30):
    km=KMeans(n_clusters=i,init='k-means++', random_state=0)
    km.fit(g_datas)
    array.append(km.inertia_)
```

```python
plot.plot(range(1,30),array)
plot.title('Elbow Method')
plot.xlabel('No of clusters')
plot.ylabel('Air Quality Monitoring data')
plot.show()
```
